```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dir = "./DATA/"
```

```python
from tensorflow.keras.preprocessing import image_dataset_from_directory
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rotation_range = 15,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    zoom_range = 0.1,
    fill_mode = 'nearest'
)
```

```python
ds = datagen.flow_from_directory(
    directory = data_dir,
    target_size = (144, 144)
)
```

```python
#plt.imshow(datagen.random_transform(my_img))
number_of_images = 20# total number of images we want to generate
images_per_row = 10# max allowed images in a row
```

```python
fig, ax = plt.subplots(nrows=number_of_images//images_per_row,ncols=images_per_row,figsize=(10,2))
count = 0
for i in range(number_of_images):
    img = next(ds)[0].astype('uint8')
    img = np.squeeze(img[0])
    ax[i // images_per_row, i % images_per_row].imshow(img)
    ax[i // images_per_row, i % images_per_row].axis('off')
```
